```python
import matplotlib.pyplot as plt
# Translation: "Hey Python, bring me the plotting toolbox, I'll call it 'plt' for short"

# Let's use simple data first
names = ['Rahul', 'Priya', 'Amit']
salaries = [50000, 65000, 48000]

# Step 1: Create a blank canvas
plt.figure(figsize=(6, 6))
# Translation: "Give me a blank white board that's 10 inches wide, 6 inches tall"
# Why? So the chart isn't tiny or huge - it's readable

# Step 2: Draw the bars
plt.bar(names, salaries, color='coral')
# Translation: "Draw vertical bars where:"
#   - X-axis (bottom) = names
#   - Y-axis (height) = salaries  
#   - Color them sky blue

# Step 3: Add a title
plt.title('Salary Comparison', fontsize=16, fontweight='bold')
# Translation: "Write 'Salary Comparison' at the top in big bold letters (size 16)"

# Step 4: Label the axes
plt.xlabel('Person Name')
# Translation: "At bottom of X-axis, write 'Person'"

plt.ylabel('Salary (₹)')
# Translation: "At left of Y-axis, write 'Salary (₹)'"

# Step 5: Tilt the names so they don't overlap
plt.xticks(rotation=0)
# Translation: "Rotate the names 45 degrees (like //// instead of ||||)"

# Step 6: Make everything fit nicely
plt.tight_layout()
# Translation: "Auto-adjust spacing so nothing gets cut off"

# Step 7: SHOW IT!
plt.show()
# Translation: "Display the chart now!"
```

```python
import seaborn as sns

# Same data
names = ['Rahul', 'Priya', 'Amit']
salaries = [50000, 65000, 48000]

# Create a DataFrame (seaborn loves these)
import pandas as pd
df = pd.DataFrame({
    'Name': names,
    'Salary': salaries
})

# Seaborn bar chart - SIMPLER
sns.barplot(data=df, x='Name', y='Salary')
plt.show('Name','Salary')

# That's it! Seaborn auto-colors, auto-styles, looks professional
```

```python

# Data: How you spend your day (hours)
activities = ['Sleep', 'Work/Study', 'Entertainment', 'Meals', 'Other']
hours = [8, 8, 4, 2, 2]

plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title("My Daily Time Breakdown", fontsize=16, fontweight='bold')
# YOUR TASK: Create a pie chart
# Use: plt.pie(hours, labels=activities, autopct='%1.1f%%')
# autopct shows percentages on slices
# Add title: "My Daily Time Breakdown"
```

```python
# Data: Ages of 20 people
ages = [23, 25, 28, 22, 25, 30, 35, 28, 22, 25, 
        27, 29, 31, 24, 26, 28, 23, 25, 27, 30]

# YOUR TASK: Create a histogram
# Use: plt.hist(ages, bins=5, color='purple', edgecolor='black')
plt.hist(ages, bins=5, color='purple', edgecolor='black')
plt.title("Age distribution", fontsize=16, fontweight='bold')
# bins=5 means "divide into 5 groups"
# edgecolor adds black borders to bars
# Title: "Age Distribution"
```
