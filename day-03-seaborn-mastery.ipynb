```python
# Import everything we need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load Titanic data (we'll use this for practice)
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

print("Data loaded successfully!")
print(f"Shape: {df.shape}")
df
```

```python
# Let's see the difference themes make

# First, create a simple chart with DEFAULT style
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Default Style - No Theme')
plt.show()
```

```python
# Now let's try different themes

# Theme 1: Darkgrid (dark background, white grid)
sns.set_style('darkgrid')
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Theme: Darkgrid')
plt.show()

# Theme 2: Whitegrid (white background, grey grid)
sns.set_style('whitegrid')
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Theme: Whitegrid')
plt.show()

# Theme 3: Dark (all dark, no grid)
sns.set_style('dark')
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Theme: Dark')
plt.show()

# Theme 4: White (clean white, no grid)
sns.set_style('white')
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Theme: White')
plt.show()

# Theme 5: Ticks (minimal, only ticks)
sns.set_style('ticks')
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Fare')
plt.title('Theme: Ticks')
plt.show()
```

```python
# EXAMPLE: Does Fare affect Survival?

# Set style
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

# Scatter plot WITH regression line
sns.regplot(data=df, x='Fare', y='Survived', 
            scatter_kws={'alpha':0.3}, 
            line_kws={'color':'red', 'linewidth':2})

plt.title('Relationship: Fare vs Survival', fontsize=16, fontweight='bold')
plt.xlabel('Fare Paid (£)', fontsize=12)
plt.ylabel('Survived (0=No, 1=Yes)', fontsize=12)
plt.ylim(-0.1, 1.1)

plt.show()
```

# Higher the fare , higher the survival chances it was

```python
# EXAMPLE 3: Does Age affect Survival rate?

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

sns.regplot(data=df, x='Age', y='Survived',
            scatter_kws={'alpha':0.3, 'color':'green'},
            line_kws={'color':'Purple', 'linewidth':2})

plt.title('Age vs Survival Rate', fontsize=16, fontweight='bold')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Survived 0=No, 1= Yes', fontsize=12)

plt.show()
```

# as the age increased the survival chances decreased

```python
# First, let's create a "correlation matrix"
# This shows: "How related is each column to each other column?"

# Select only numeric columns
numeric_cols = df[['Age', 'Fare', 'Survived', 'Pclass']].corr()

print("Correlation matrix:")
print(numeric_cols)
```

```python
# Create heatmap of correlations

sns.set_style('white')
plt.figure(figsize=(8, 6))

# Create the heatmap
sns.heatmap(numeric_cols, 
            annot=True,           # Show numbers in cells
            cmap='coolwarm',      # Color scheme (blue=cold, red=hot)
            center=0,             # Center the colors at 0
            square=True,          # Make cells square
            linewidths=1,         # Lines between cells
            cbar_kws={'label': 'Correlation'})

plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.show()
```

```python
# Create a crosstab (counts) of survival by class and gender
survival_table = pd.crosstab([df['Pclass'], df['Sex']], df['Survived'])

print("Survival counts:")
print(survival_table)
print()

# Now visualize it
plt.figure(figsize=(8, 6))

sns.heatmap(survival_table, annot=True, fmt='d', cmap='YlGnBu', 
            cbar_kws={'label': 'Number of Passengers'})

plt.title('Survival Counts by Class and Gender', fontsize=16, fontweight='bold')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Class & Gender')
plt.show()
```

```python
# Import everything
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Set style
sns.set_style('whitegrid')

# THE MAGIC LINE - Create a 2x2 grid of charts
fig, axes = plt.subplots(3, 2, figsize=(16, 12))

# This creates 4 empty "slots" for charts:
# [0,0] [0,1]
# [1,0] [1,1]

# CHART 1 (Top-Left): Survival Count
sns.countplot(data=df, x='Survived', palette='pastel', ax=axes[0, 0])
axes[0, 0].set_title('1. Survival Count', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Survived (0=No, 1=Yes)')
axes[0, 0].set_ylabel('Number of Passengers')


# CHART 2 (Top-Right): Survival Rate by Class
sns.barplot(data=df, x='Pclass', y='Survived', palette='viridis', ax=axes[0, 1])
axes[0, 1].set_title('2. Survival Rate by Passenger Class', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Passenger Class')
axes[0, 1].set_ylabel('Survival Rate')


# CHART 3 (Bottom-Left): Age Distribution by Survival
sns.boxplot(data=df, x='Survived', y='Age', palette='coolwarm', ax=axes[1,0])
axes[1, 0].set_title('3 Age Distribution by Survival', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Survived')
axes[1, 0].set_ylabel('Age')

# CHART 4 (Bottom-Right): Age Distribution by Survival
# First create the crosstab
survival_pivot = df.pivot_table(values='PassengerId', 
                                 index='Pclass', 
                                 columns='Survived', 
                                 aggfunc='count')

sns.heatmap(survival_pivot, annot=True, fmt='g', cmap='Blues', 
            ax=axes[1, 1], cbar_kws={'label': 'Count'})
axes[1, 1].set_title('4. Survival Count by Class', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Survived (0=No, 1=Yes)')
axes[1, 1].set_ylabel('Passenger Class')

# CHART 5: Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set1', ax=axes[2, 0])
axes[2, 0].set_title('5. Survival by Gender', fontsize=14, fontweight='bold')
axes[2, 0].set_xlabel('Gender')
axes[2, 0].set_ylabel('Count')

# CHART 5: Pclass by Gender
sns.countplot(data=df, x='Sex', hue='Pclass', palette='Set1', ax=axes[2, 1])
axes[2, 1].set_title('6. Survival by Gender', fontsize=14, fontweight='bold')
axes[2, 1].set_xlabel('Gender')
axes[2, 1].set_ylabel('Count')

# Add overall title for the entire dashboard
fig.suptitle('TITANIC SURVIVAL ANALYSIS DASHBOARD', 
             fontsize=20, fontweight='bold', y=0.995)

# Adjust spacing so charts don't overlap
plt.tight_layout()

# Display the dashboard!
plt.show()
```
